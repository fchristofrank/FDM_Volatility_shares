```python
import numpy as np
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import math

# n,m are the dimensions of the array containing the space and time
n = 20
m = 20
x,time = np.meshgrid(np.linspace(0,2,n),np.linspace(0,1,m))

#Initialize the volatility matrix
V = np.zeros((n,m))
for i in range(n):
    V[i][0] = max(math.exp(i)-1,0)

#Define various parameters required for the FDM
dt = 0.0001
h = 2/20
#where 'k' is the constant in the equation
k = 2

for t in range(1,n):
    for i in range(2,m):
        V[i][t] = V[i][t-1] + dt*((V[i][t-1] - 2*V[i-1][t-1] +V[i-2][t-1])/h**2 + ((k-1)/h)*(V[i][t-1] - V[i-1][t-1]) - k*V[i][t-1])
        #print(V[i][t])
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x,time, V, cmap=cm.coolwarm,linewidth=0.5, antialiased=False)   
 
        
        
```
